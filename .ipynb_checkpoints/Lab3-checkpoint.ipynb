## Problem 2
In this example we will be looking for asteroids. If we look at the alignment of stars on subsequent images, they don't perfectly align due to atmospheric and instrumental effects (even ignoring proper motion). The resulting distribution is two dimensional, and for this lab let's assume it is a 2D Gaussian with 1 arcsecond RMS. Or said another way, if I histogram how far all the (stationary) stars appear to have moved I get something like:

```python
a = np.vstack((stats.norm.rvs( scale = 1, size = 100000), stats.norm.rvs( scale = 1, size = 100000)))
a.shape
fig, ax = plt.subplots(1, 1)
h = ax.hist2d(a[0,:],a[1,:],bins=100, density=True);
ax.set_aspect('equal', 'box')
plt.xlim([-3 , 3])
plt.ylim([-3 , 3])
plt.title("2D Histogram of positional uncertainty", fontsize = 24)
plt.ylabel("$\Delta$y arcseconds", fontsize = 18)
plt.xlabel("$\Delta$x arcseconds", fontsize = 18)
plt.colorbar(h[3], ax=ax)
```

If I have a potential asteroid, it will have some true movement between the images. We would like a '5 sigma' detection of movement. What is that distance in arcseconds?

#### 1. What is the question in words?
If there is no signal, or if there are no asteroids detected in these images, then the background data is how far the stationary stars appear to have moved and this data has a 2D Guassian distribution with a standard deviation of one arcsecond. What is the probability that the data produces a signal greater than 5 sigma, which is our threshold for differentiating true movement of asteroids from the movement generated by instrumental and environmental uncertainty of stationary stars, and what distance in arcseconds does that probability correspond to?

#### 2. What is the question in math? [Hint: Come up with an analytic description of your pdf(). The Blue Book may be helpful here.]
The distance can be described as the amplitude of the vector created by these two 1D Guassian distributions. As we found in HW3, this amplitude vector has a rayleigh distribution. In math, we want to find this pdf, then given the probabilty of 5 sigma, we want to find the area under the rayleigh distribution that corresponds to 5 sigma, then find the value (x) on the rayleigh distribution that corresponds to this probability. This x value will be our 5 sigma detection of movement in arcseconds.

As we can see from the plot and log plot below, the data closely follows a rayleigh distribution.

```python
amplitude = np.sqrt(a[0,:]**2 + a[1,:]**2)
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

ax[0].hist(amplitude, bins=100, density=True)
x = np.linspace(0, 5, 1000)
ax[0].plot(x, stats.rayleigh.pdf(x), linewidth = 5, alpha = 0.7)
ax[0].set_title("Amplitude of Alignment of Stars", fontsize=18)
ax[0].set_xlabel("Distance in Arcseconds", fontsize=18)
ax[0].set_ylabel("Frequency", fontsize=18)
plt.tick_params(labelsize = 18)

ax[1].hist(amplitude, bins=100, density=True)
x = np.linspace(0, 5, 1000)
ax[1].plot(x, stats.rayleigh.pdf(x), linewidth = 5, alpha = 0.7)
ax[1].set_title("Log of Amplitude of Alignment of Stars", fontsize=18)
ax[1].set_xlabel("Distance in Arcseconds", fontsize=18)
ax[1].set_ylabel("Log Frequency", fontsize=18)
ax[1].set_yscale('log')
```

#### 3. What distance in arcseconds represents a 5 sigma detection of motion?

```python
# find the probability associated with 5 sigma - under the standard normal, then find the value corresponding to that probability
# on our Rayleigh distribution
prob_5_sigma = 1 - stats.norm.cdf(5)
print("Probability associated with 5 sigma: " + str(prob_5_sigma))
distance = stats.rayleigh.ppf(stats.norm.cdf(5))
print("Distance in arcseconds that represents 5 sigma detection of motion: " + str(distance))
```

## Problem 3
As we discussed in class, one of the key backgrounds for gamma-ray telescopes are cosmic rays. Cosmic rays are charged particlesusually protons or electrons but can include atomic nuclei such a alpha particles (helium) or iron. Because of their charge cosmic rays spiral in the magnetic field of the galaxy. From the perspective of the Earth they appear to be coming uniformly from all directions like a high energy gas, and the direction the cosmic ray is travelling when it reaches the Earth tells us nothing about where it came from because we don't know what tortured path it has taken through the galaxy to reach us. However, at trillion electron volt energies and above, the spiral loops are fairly big and the sun and the moon will block cosmic rays. This means the sun and the moon appear as holes in the cosmic ray sky (cosmic rays from that direction are absorbed).

Assume in a moon sized patch on the sky we normally have a cosmic ray rate of 1 cosmic ray per minute (arrivals are random in time). If we can observe where the moon is for 8 hours per night (not too close to the horizon) and we observe for 15 days and see 6800 cosmic rays, what is the signficance of our moon shadow detection?

As always:

#### 1. What is the question in words?

Given that the data of cosmic rays over 15 days at 8 hours per night with 1 cosmic ray per minute follows a Poisson distribution with expected value of 7200, what is the probabilty of getting the value 6800 cosmic rays over this time period?

#### 2. Translate the question to math, describing both the expected background pdf() and the integral you need to do.

The expected background pdf is a Poisson distribution with expected value of 7200. We need to find the probability of exactly 6800 from this background distribution, then do the integral under the standard normal distribution that has this same probability to find the sigma.

#### 3. Convert to sigma. 

```python
fig, ax = plt.subplots(1, 1)
mu = 7200
x = np.arange(stats.poisson.ppf(0.001, mu), stats.poisson.ppf(0.999, mu))
ax.step(x, stats.poisson.pmf(x, mu))
plt.title("Poisson Distribution of Cosmic Rays in Moon Shadow", fontsize=18)
plt.xlabel("Number of Cosmic Rays", fontsize=18)
plt.ylabel("Probability", fontsize=18)
```

```python
prob_6800 = stats.poisson.pmf(6800, mu)
print("Probability of getting 6800 cosmic rays in a moon-shaped patch of sky over 15 days measuring 8 hours a night: " + str(prob_6800))
sigma_6800 = stats.norm.ppf(prob_6800)
print("The associated sigma value: "+ str(abs(sigma_6800)))
```

#### There are some unique aspects to this problem. Make sure you explore them in your explanation.

Some unique aspects to this problem are that since that data is number of occurences over a certain period of time, the data is a discrete Poisson distribution instead of a continuous one. Since the data follows a Poisson distribution, instead of taking the probabilty of an area, we are only interested in the probability of getting 6800 exactly, and then finding the corresponding area under the standard normal to find the sigma. The sigma also came out negative at first because the probability we got from the Poisson distribution is close to zero, so the absolute value of the negative sigma is the sigma value we want.
