# Problem 2
#### 2a) What is the pdf of the sum of two identical exponential distributions?

```python
# 2. Analytical solutions; applying mathematical operations to probability distributions
# 2a. Sum of two idential exponential distributions
d = stats.expon.rvs(size = 100000)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20,6))
ax[0].hist(d,50, density=True)
ax[0].set_title("Exponential Distribution PDF", fontsize=18)
ax[0].set_xlabel("Values", fontsize=18)
ax[0].set_ylabel("Frequency", fontsize=18)
x = np.linspace(0,10,1000)
ax[0].plot(x,stats.expon.pdf(x, scale=1),linewidth = 8,alpha = 0.7)

ax[1].hist(d+d,50, density=True)
ax[1].set_title("Sum of Two Identical Distributions PDF", fontsize=18)
ax[1].set_xlabel("Values", fontsize=18)
ax[1].set_ylabel("Frequency", fontsize=18)
x = np.linspace(0,20,1000)
ax[1].plot(x,stats.expon.pdf(x, scale=2),linewidth = 8,alpha = 0.7)

fig, ax = plt.subplots(1, 1)
x = np.linspace(0,10,1000)
ax.plot(x,stats.expon.pdf(x, scale=1),linewidth = 3,alpha = 0.7, label="One Exponential Distribution")
x = np.linspace(0,20,1000)
ax.plot(x,stats.expon.pdf(x, scale=2),linewidth = 3,alpha = 0.7, label="Sum of Two Identical Exponential Distributions")
plt.title("Curves Representing both PDF", fontsize=18)
plt.xlabel("X Values", fontsize=18)
plt.ylabel("Y values", fontsize=18)
plt.legend()
```

#### 2a) Judging from the graphs, the pdf of the sum of two idential exponential distributions is another exponential distribution with lambda=0.5. The scale of the plot for the sum distribution is scale=2, which according to the stats.expon documentation means that the parameter lambda = 0.5

#### 2b) What is the pdf of the ratio of two zero-mean unity variance normal distributions  X1/X2 ?

```python
a = stats.norm.rvs(loc = 0, scale = 1, size = 5000)
b = stats.norm.rvs(loc = 0, scale = 1, size = 5000)
fig, ax = plt.subplots(1, 1)
ax.hist(a,100, density=True)
ax.hist(b,100, density=True)
plt.title("Zero-Mean Unity Variance Normal Distribution PDF", fontsize=18)
plt.xlabel("Values", fontsize=18)
plt.ylabel("Frequency", fontsize=18)

fig, ax = plt.subplots(1, 1)
ax.hist(a/b, 5000, density=True, label="Ratio of Two Normal Distributions")
plt.xlim(-5, 5)
x = np.linspace(-5, 5, 1000)
ax.plot(x, stats.cauchy.pdf(x), linewidth = 5, alpha = 0.7, label="Cauchy Distribution")
plt.title("Ratio of Two Normal Distributions PDF", fontsize=18)
plt.xlabel("Values", fontsize=18)
plt.ylabel("Frequency", fontsize=18)
plt.legend()
```

#### The ratio of two zero-mean unity variance normal distributions is a Cauchy distribution according to Wikipedia. Plotting the ratio distribution supports this conclusion.

#### 2c) So far we have looked at 1D probability distributions, but it is possible to have a multi-dimensional vector distribution. A simple first introduction is the 2D Guassian; it looks like a smeared spot. Mathematically this is given by  Xi^+Yj^  where both  X  and  Y  are drawn from 1D Gaussian distributions. If I measure the amplitude of this vector, what is its pdf? (Hint, the amplitude is always positive.)

```python
a = stats.norm.rvs(loc = 0, scale = 1, size = 100000)
b = stats.norm.rvs(loc = 0, scale = 1, size = 100000)

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
plt.title("2D graph of 2D Gaussian", fontsize=18)
plt.xlabel("First Normal Distribution", fontsize=18)
plt.ylabel("Second Normal Distribution", fontsize=18)
ax.plot(a, b)

amplitude = np.sqrt(a**2 + b**2)
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.hist(amplitude, bins=100, density=True)
x = np.linspace(0, 5, 1000)
ax.plot(x, stats.rayleigh.pdf(x), linewidth = 5, alpha = 0.7, label="Rayleigh Distribution")
plt.title("Amplitude of 2D Gaussian Vector", fontsize=18)
plt.xlabel("Values", fontsize=18)
plt.ylabel("Frequency", fontsize=18)
plt.legend()
```

#### After graphing the amplitude of the vector that is the sum of the two Guassian distributions, the resulting distribution closely matches that of the Rayleigh probability distribution function with sigma=1.
